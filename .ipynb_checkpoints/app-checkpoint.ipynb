```python
from langchain_core.prompts import PromptTemplate, load_prompt
from PromptManager import PromptManager
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableParallel, RunnablePassthrough
import os
```

```python
topic_creator_prompt = load_prompt(PromptManager.TOPIC_CREATOR)
post_generator_prompt = load_prompt(PromptManager.POST_GENERATOR)
hashtag_suggestor_prompt = load_prompt(PromptManager.HASHTAG_SUGGESTOR)
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", google_api_key=os.getenv("GOOGLE_API_KEY"), verbose=True)
output_parser = StrOutputParser()
```

```python
tone = "Pragmático, direto e humano"
topic = "Migração WebForms para MVC"

topic_chain = topic_creator_prompt | llm | output_parser
post_chain = (
    {"contexto": RunnablePassthrough(), "tom_de_voz": lambda _: tone}
    | post_generator_prompt
    | llm
    | StrOutputParser()
)

hashtag_chain = (
    {"conteudo_post": RunnablePassthrough()}
    | hashtag_suggestor_prompt
    | llm
    | output_parser
)
combined_chain = post_chain | RunnableParallel(
    post=RunnablePassthrough(),
    hashtags=hashtag_chain
)

chain = topic_chain | combined_chain

result = chain.invoke({"tema": topic})

print("--- POST GERADO ---")
print(result['post'])
print("\n--- HASHTAGS ---")
print(result['hashtags'])
```
